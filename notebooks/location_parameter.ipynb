```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf as pdf
import os
```

```python
folder_unprep = "/Users/harun/Desktop/Uni/4. Semester/Data Mining/Projekt/data (bearbeitet)"
folder_prep = "/Users/harun/Desktop/Uni/4. Semester/Data Mining/GroupA_DataMining/resources/data-prep"
sub_folder1 = "SCRIPTURES"
sub_folder2 = "csv-files"
csv_folder_path_unprep = f"{folder_unprep}/{sub_folder1}/{sub_folder2}"
csv_folder_path_prep = folder_prep

# Find all datasets
cwd = os.getcwd()
all_datasets = os.listdir(csv_folder_path_prep)
print(type(all_datasets))
```

```python
GREEN = "#3AB795"
YELLOW = "#FFCF56"
RED = "#8C1C13"
```

```python
# Calculate arithmetic mean of every feature of every dataset
def arithmetic_mean():
    pdf_file = pdf.PdfPages('arithmetic_means_prep(min_max).pdf')
    
    for filename in all_datasets:
        if filename.endswith(".csv"):
            df = pd.read_csv(f"{csv_folder_path_prep}/{filename}")
            row_count = df.shape[1]
            mean_features = []
            
            plt.figure()
            
            for i in range(row_count):
                mean_features.append(df.iloc[:, i].mean())

            index = range(1, len(mean_features)+1)

            bars = plt.bar(index, mean_features)
            plt.title(filename[:-4])
            plt.xlabel("Feature")
            plt.ylabel("Arithmetic Mean")
            
            plt.xticks(index, rotation=90)
            plt.tight_layout()
            
            for bar in bars:
                if round(bar.get_height(), 2) == 0:
                    bar.set_facecolor(YELLOW)
                    plt.text(bar.get_x() + bar.get_width() / 2, 0, "0", ha='center', va='bottom')
                else:
                    bar.set_facecolor(GREEN)
            
            pdf_file.savefig(plt.gcf(), bbox_inches='tight')
            
    pdf_file.close()
arithmetic_mean()
```

```python
def min_max():
    pdf_file = pdf.PdfPages('min_max_prep(min_max).pdf')
    
    for filename in all_datasets:
        if filename.endswith(".csv"):
            df = pd.read_csv(f"{csv_folder_path_prep}/{filename}")
            row_count = df.shape[1]
            min_features = []
            max_features = []
            
            plt.figure()
            
            for i in range(row_count):
                min_features.append(df.iloc[:, i].min())
                max_features.append(df.iloc[:, i].max())
            index = range(1, len(min_features)+1)
            
            print(min_features)
            print(max_features)
            
            for min_y, max_y, idx in zip(min_features, max_features, index):
                plt.bar(idx, max_y, color=GREEN)
                plt.bar(idx, min_y, color=YELLOW)
            
            
            plt.title(filename[:-4])
            plt.xlabel("Feature")
            plt.ylabel("Min_Max")
            plt.xticks(index, rotation=90)
            plt.legend(['Max', 'Min'], loc=1)
            plt.tight_layout()
            
            pdf_file.savefig(plt.gcf(), bbox_inches='tight')   
    pdf_file.close()
min_max()
```

```python
def variance():
    pdf_file = pdf.PdfPages('variance_prep(min_max).pdf')
    
    for filename in all_datasets:
        if filename.endswith('.csv'):
            df = pd.read_csv(f"{csv_folder_path_prep}/{filename}")
            row_count = df.shape[1]
            variances_features = []
            
            plt.figure()
            
            for i in range(row_count):
                variances_features.append(df.iloc[:, i].var())
            
            std_of_variances = np.std(variances_features)
            
            index = range(1, len(variances_features)+1)
            
            bars = plt.bar(index, variances_features)
            plt.xlabel("Feature")
            plt.ylabel("Variances")
            plt.title(filename[:-4])
            plt.xticks(index, rotation=90)
            
            for bar in bars:
                if bar.get_height() > std_of_variances:
                    bar.set_facecolor(RED)
                else:
                    bar.set_facecolor(GREEN)
            
            pdf_file.savefig(plt.gcf(), bbox_inches='tight')
    pdf_file.close()

variance()
```
